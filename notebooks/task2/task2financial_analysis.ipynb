```python
import pandas as pd
import os

# Define the path to the data directory
data_directory = 'c:/Users/Yibabe/Desktop/kifiyaweek1/data'

# List of CSV filenames
csv_files = [
    'AAPL_historical_data.csv',
    'AMZN_historical_data.csv',
    'GOOG_historical_data.csv',
    'META_historical_data.csv',
    'MSFT_historical_data.csv',
    'NVDA_historical_data.csv',
    'TSLA_historical_data.csv'
]
# Create a dictionary to hold DataFrames
dfs = {}

# Load CSV files into DataFrames
for file in csv_files:
    file_path = os.path.join(data_directory, file)
    dfs[file] = pd.read_csv(file_path)

# Check the loaded DataFrames
for name, df in dfs.items():
    print(f"{df.head()}")

```

```python
# Check for missing values
for name, df in dfs.items():
    print(f"{name} missing values:")
    print(df.isnull().sum())
```

```python
# Drop rows with missing values (if appropriate)
for name, df in dfs.items():
    dfs[name] = df.dropna()
```

```python
# Convert date columns to datetime (if applicable)
for name, df in dfs.items():
    if 'Date' in df.columns:
        dfs[name]['Date'] = pd.to_datetime(dfs[name]['Date'], errors='coerce')
```

```python
import talib
import pandas as pd

# Assuming dfs is a dictionary of DataFrames
for name, df in dfs.items():
    # Check if 'Close' column exists in the DataFrame
    if 'Close' in df.columns:
        # Calculate Moving Averages
        df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)
        df['SMA_200'] = talib.SMA(df['Close'], timeperiod=200)

        # Calculate RSI
        df['RSI'] = talib.RSI(df['Close'], timeperiod=14)

        # Calculate MACD
        df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)

        # Print to check calculations
        print(f"{name} Technical Indicators:")
        print(df[['Date', 'Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal', 'MACD_hist']].head())
    else:
        print(f"{name} does not have 'Close' column. Skipping indicator calculations.")
```
