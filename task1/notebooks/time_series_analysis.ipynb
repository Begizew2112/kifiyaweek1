```python
import pandas as pd
# Load the preprocessed data
df = pd.read_csv('../data/processed_analyst_ratings.csv', parse_dates=['date'])
df.set_index('date', inplace=True)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Resample to daily frequency and count the number of articles per day
daily_counts = df.resample('D').size()

# Plot the number of articles by day
plt.figure(figsize=(12, 6))
plt.plot(daily_counts.index, daily_counts.values, label='Daily Count', color='blue')
plt.title('Number of Articles Published Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Resample to monthly frequency and count the number of articles per month
monthly_counts = df.resample('M').size()

# Plot the number of articles by month
plt.figure(figsize=(12, 6))
plt.plot(monthly_counts.index, monthly_counts.values, label='Monthly Count', color='green')
plt.title('Number of Articles Published Per Month')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extract the hour of the day from the index
df['hour'] = df.index.hour

# Plot the number of articles by hour of the day
plt.figure(figsize=(10, 5))
sns.histplot(df['hour'], bins=24, kde=False, color='purple')
plt.title('Number of Articles Published by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Articles')
plt.grid(True)
plt.show()
```

```python
# Compute rolling mean and rolling standard deviation
rolling_mean = monthly_counts.rolling(window=12).mean()
rolling_std = monthly_counts.rolling(window=12).std()

# Plot the monthly counts with rolling statistics
plt.figure(figsize=(12, 6))
plt.plot(monthly_counts.index, monthly_counts.values, label='Monthly Count', color='green')
plt.plot(rolling_mean.index, rolling_mean.values, label='Rolling Mean', color='red')
plt.plot(rolling_std.index, rolling_std.values, label='Rolling Std Dev', color='orange')
plt.title('Monthly Count with Rolling Mean and Std Dev')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.legend()
plt.grid(True)
plt.show()
```
